```python
# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])

# vocabulary size
total_words = len(tokenizer.word_index) + 1 # +1 for the OOV token (out of vocabulary) or index 0
print(f"Total unique words (vocabulary size): {total_words}")

# Create input sequences
input_sequences = []
sentences = text.replace('\n', ' ').split('.')
for sentence in sentences:
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]
    if len(tokenized_sentence) > 1:
        for i in range(1, len(tokenized_sentence)):
            input_sequences.append(tokenized_sentence[:i+1])

print(f"Total input sequences created: {len(input_sequences)}")

# Pad sequences
max_len = max(len(x) for x in input_sequences)
print(f"Maximum sequence length: {max_len}")

padded_input_sequence = pad_sequences(input_sequences, maxlen=max_len, padding='pre')
print(f"Shape of padded input sequences: {padded_input_sequence.shape}")

# Split into X and Y
X = padded_input_sequence[:, :-1]
y = padded_input_sequence[:, -1]

print(f"Shape of X (input features): {X.shape}")
print(f"Shape of y (target labels): {y.shape}")

# One-hot encode Y
y = to_categorical(y, num_classes=total_words)
print(f"Shape of y after one-hot encoding: {y.shape}")

# Build the LSTM model
model = Sequential()

model.add(Embedding(total_words, 150, input_length=max_len - 1))
model.add(LSTM(500))
model.add(Dense(total_words, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

print("\nModel Summary:")
model.summary()

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Train the model
print("\nStarting model training...")
history = model.fit(X, y, epochs=100, verbose=1).to(device)

print("\nTraining complete.")
```

```python
# Next Word Prediction Function

def predict_next_words(seed_text, num_words_to_generate, model, tokenizer, max_sequence_len):
    generated_text = seed_text
    print(f"\nSeed Text: {seed_text}")
    print("Generating...")

    for _ in range(num_words_to_generate):
        # Tokenize the current generated_text
        token_list = tokenizer.texts_to_sequences([generated_text])[0]

        # Pad the tokenized sequence
        padded_token_list = pad_sequences([token_list], maxlen=max_sequence_len - 1, padding='pre')

        # Predict the probabilities for the next word
        predicted_probabilities = model.predict(padded_token_list, verbose=0)[0]

        # Get the index of the word with the highest probability
        predicted_word_index = np.argmax(predicted_probabilities)

        # Find the word corresponding to the index
        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted_word_index:
                output_word = word
                break

        if output_word == "":
            break

        generated_text += " " + output_word
        # print(generated_text)
        time.sleep(0.1)

    return generated_text
```
